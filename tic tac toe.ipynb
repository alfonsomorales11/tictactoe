```python
def show_board(board):
    print(board[0])
    print(board[1])
    print(board[2])
    print('')
```

```python
def init_board():
    return [['1','2','3'],['4','5','6'],['7','8','9']]
```

```python
def select_position(player_number,positions_list,winner_flag_string):
    
    is_a_digit = False
    in_position_range = False
    is_position_taken = True
    
    symbol = ''
    if ( player_number == 1 ):
        symbol = 'X'
        
    else:
        symbol = 'O'
    
    while ( is_a_digit == False or in_position_range == False or is_position_taken == True ):
        
        if ( len(positions_list) == 9 ):
            print('All the positions are taken!')
            position = -1
            break
            
        print('[    Player {} turn ({})    ]'.format(player_number,symbol))
        position = input('Select a position (1-9): ')
        
        if ( position.isdigit() ):
                position = int(position)
                is_a_digit = True
                
                if ( position in range (1,10) ):
                    in_position_range = True
                    
                    if ( position not in positions_list ):
                        positions_list.append(position)
                        is_position_taken = False
                        
                    else:
                        print('That position is already taken\n')
                
                else:
                    print('The number is not within range (1,9)\n')
                    
        else:
            print('Not a number, try again')
    
    return position
```

```python
def place_selection(position,board,player_number):
    
    position = position -1
    
    if ( position >= 0 and position < 3 ):
        if ( player_number == 1 ):
            board[0][position] = "X"
            
        elif ( player_number == 2 ):
            board[0][position] = "O"
        
    elif ( position >=3 and position < 6 ):
        if ( player_number == 1 ):
            board[1][position-3] = "X"
            
        elif ( player_number == 2 ):
            board[1][position-3] = "O"
        
    elif ( position >=6 and position < 9 ):
        if ( player_number == 1 ):
            board[2][position-6] = "X"
            
        elif ( player_number == 2 ):
            board[2][position-6] = "O"
    
    return board
```

```python
def check_winner(boardx):
    
    winner_string = ''
        
    # Player 1 (X)
    if ( boardx[0].count('X') == 3 or 
         boardx[1].count('X') == 3 or 
         boardx[2].count('X') == 3 or
         boardx[0][0] == 'X' and boardx[1][0] == 'X' and boardx[2][0] == 'X' or
         boardx[0][1] == 'X' and boardx[1][1] == 'X' and boardx[2][1] == 'X' or
         boardx[0][2] == 'X' and boardx[1][2] == 'X' and boardx[2][2] == 'X' or
         boardx[0][0] == 'X' and boardx[1][1] == 'X' and boardx[2][2] == 'X' or
         boardx[2][0] == 'X' and boardx[1][1] == 'X' and boardx[0][2] == 'X'
        ):
        
        winner_string = "\nPlayer 1 wins\n"
        
    # Player 2 (O)
    elif ( boardx[0].count('O') == 3 or
           boardx[1].count('O') == 3 or
           boardx[2].count('O') == 3 or
           boardx[0][0] == 'O' and boardx[1][0] == 'O' and boardx[2][0] == 'O' or
           boardx[0][1] == 'O' and boardx[1][1] == 'O' and boardx[2][1] == 'O' or
           boardx[0][2] == 'O' and boardx[1][2] == 'O' and boardx[2][2] == 'O' or
           boardx[0][0] == 'O' and boardx[1][1] == 'O' and boardx[2][2] == 'O' or
           boardx[2][0] == 'O' and boardx[1][1] == 'O' and boardx[0][2] == 'O'
         ):
        
        winner_string = "\nPlayer 2 wins\n"
    
    else:
        pass
    
    return winner_string
```

```python
def print_actual_board (board):
    uno = '       |       |'
    dos = '-------+-------+------'

    uno_nums  = '   1   |   2   |   3'
    dos_nums  = '   4   |   5   |   6'
    tres_nums = '   7   |   8   |   9'
    
    uno_nums = list(uno_nums)
    dos_nums = list(dos_nums)
    tres_nums = list(tres_nums)

    word1 = ''
    word2 = ''
    word3 = ''
    
    indexes_to_change = [3,11,19]
    
    for i in range (len(indexes_to_change)):
        uno_nums[indexes_to_change[i]] = board[0][i]
        dos_nums[indexes_to_change[i]] = board[1][i]
        tres_nums[indexes_to_change[i]] = board[2][i]
        
    for j in uno_nums:
        
        word1 = word1 + j
        
    for k in dos_nums:
        word2 = word2 + k
        
    for n in tres_nums:
        word3 = word3 + n
        
    print('')
    print(uno)
    print(word1)
    print(uno)
    print(dos)
    print(uno)
    print(word2)
    print(uno)
    print(dos)
    print(uno)
    print(word3)
    print(uno)
    print('')
```

```python
# Start a new game
board = init_board()
        
# Initialize variables
taken_positions = []
winner_flag = ''
new_game_option = True

while ( len(winner_flag) == 0 ):
                   
    if ( winner_flag == '' ):
        # Show board
        print_actual_board(board)
        
        # Ask for position player 1
        player1_position = select_position(1,taken_positions,winner_flag)

        # Place token in board - player 1
        board = place_selection(player1_position,board,1)

        # Check if player 1 has won
        winner_flag = check_winner(board)
        print(check_winner(board))
    
    if ( winner_flag == '' ):
        # Show board
        print_actual_board(board)
        
        # Ask for position player 2
        player2_position = select_position(2,taken_positions,winner_flag)

        # Place token in board - player 2
        board = place_selection(player2_position,board,2)

        # Check if player 2 has won
        winner_flag = check_winner(board)
        print(check_winner(board))
        
    if ( winner_flag != '' ):
        print_actual_board(board)
        
        
        while ( new_game_option == True ):
            new_game = input('Do you want to play again?: (yes/no)')
            
            if ( new_game.lower() == 'yes' or new_game.lower() == 'no' ):
                new_game_option =  False
            
            else:
                print('You wrote an invalid option')
        
        if ( new_game == 'yes' ):
            # Clean the board
            board = init_board()
            
            # Re-initialize variables
            taken_positions = []
            winner_flag = ''
            new_game_option = True
            continue
        
        else:
            print('See you later')
            break   
```
